```python
import pandas as pd

df = pd.read_csv('네이버웹툰_최종_프로젝트.csv')
df
```

```python
import pickle

with open('similarities_resnet.pkl', 'rb') as f:
    similarities_resnet = pickle.load(f)

with open('similarities_st.pkl', 'rb') as f:
    similarities_st = pickle.load(f)

with open('similarities_wv.pkl', 'rb') as f:
    similarities_wv = pickle.load(f)
```

```python
## 정규화o (정규화 x가 나은듯!!)

import numpy as np

def normalize_similarity_matrix(matrix):
    min_val = np.min(matrix)
    max_val = np.max(matrix)
    normalized_matrix = (matrix - min_val) / (max_val - min_val)
    return normalized_matrix

normalized_resnet = normalize_similarity_matrix(similarities_resnet)
normalized_st = normalize_similarity_matrix(similarities_st)
normalized_wv = normalize_similarity_matrix(similarities_wv)
```

```python
final_similarities = normalized_resnet * 0.6 + normalized_st * 0.2 + normalized_wv * 0.2
```

```python
## 정규화x

import numpy as np

# numpy 배열로 변환
similarities_resnet = np.array(similarities_resnet)
similarities_st = np.array(similarities_st)
similarities_wv = np.array(similarities_wv)
```

```python
final_similarities = similarities_resnet * 1 + similarities_st * 0 + similarities_wv * 0
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import warnings

# 경고 무시
warnings.filterwarnings('ignore')

# 한글 깨짐 방지
plt.rc('font', family='NanumBarunGothic')

# 웹툰명에 해당하는 인덱스를 찾아 해당 인덱스의 유사도 리스트 가져오기
webtoon_name = input("웹툰명을 입력하세요: ")
index = df[df['제목'] == webtoon_name].index[0]
similarity_list = final_similarities[index]

# 유사도 리스트 numpy로 변환하고 유사도 상위 10개 index 가져오기
similarity_array = np.array(similarity_list)
top_indices = similarity_array.argsort()[-10:][::-1]

# 상위 10개 웹툰 df 따로 만들기
display_df = df.loc[list(top_indices)]
display_df['유사도'] = display_df.index.map(lambda x: similarity_array[x] if x != index else '기준')

# 사용자가 입력한 웹툰과의 유사도 상위 10개 출력(이미지)
fig, axs = plt.subplots(2, 5, figsize=(15, 6))

for idx, i in enumerate(top_indices):
    row = idx // 5
    col = idx % 5

    title = df.loc[i, '제목']
    similarity = similarity_array[i] if i != index else '기준'

    # 이미지 파일 경로 생성
    content_link = df.loc[i, '내용링크']
    unique_id = content_link.split('=')[-1]
    index_str = str(i).zfill(4)
    img_path = f"./naver_webtoon_thumbnail_first/{index_str}_{unique_id}.jpg"

    # 이미지를 로드&출력
    img = mpimg.imread(img_path)
    axs[row, col].imshow(img)
    axs[row, col].set_title(f"{title}\n유사도: {similarity}")
    axs[row, col].axis('off')

plt.tight_layout()
plt.show()
```

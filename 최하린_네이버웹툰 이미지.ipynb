```python
import requests
import os
from tqdm.notebook import tqdm

def download_image(img_url, idx, save_folder='downloaded_images'):
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    # 이미지링크에 주소 입력 안 될 시 버림
    if not img_url.startswith('http'):
        return

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    response = requests.get(img_url, headers=headers, stream=True)
    response.raise_for_status()  # 오류 발생하면 예외

    # 이미지 이름 설정 (0001.jpg, 0002.jpg, ...)
    img_name = f"{idx:04}.jpg"

    # 이미지 저장
    with open(os.path.join(save_folder, img_name), 'wb') as file:
        for chunk in response.iter_content(1024):
            file.write(chunk)

# tqdm을 사용하여 진행 상태 표시
counter = 1
for link in tqdm(df['이미지링크']):
    download_image(link, counter)
    counter += 1
```

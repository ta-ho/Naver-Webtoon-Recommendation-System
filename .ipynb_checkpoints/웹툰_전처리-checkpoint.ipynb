```python
import pandas as pd

df = pd.read_csv("네이버웹툰_최종.csv")
df[:10]
```

```python
# 글,그림,스토리
df_two_slash = df[df['작가'].str.count('/') == 2]
```

```python
def split_author(a):
    parts = a.split("/")
    if len(parts) == 1:
        return parts[0], parts[0]
    else:
        return parts[0], parts[1]
    
df['글'], df['그림'] = zip(*df['작가'].apply(split_author))
```

```python
df_two_slash = df[df['작가'].str.count('/') == 2]
df_two_slash
```

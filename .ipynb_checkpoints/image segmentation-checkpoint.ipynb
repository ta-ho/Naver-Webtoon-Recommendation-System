## 1단계: 필요한 라이브러리와 모델 로드

```python
import torch
import torchvision.transforms as transforms
import torchvision.models as models
import torchvision.models.detection as detection
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

transform = transforms.Compose([
    #transforms.Resize((202, 120)),
    transforms.ToTensor(),
])

maskrcnn = detection.maskrcnn_resnet50_fpn(pretrained=True)
maskrcnn = maskrcnn.to(device)
maskrcnn.eval()

resnet = models.resnet50(pretrained=True).to(device)
resnet = torch.nn.Sequential(*list(resnet.children())[:-1])

class GramMatrix(torch.nn.Module):
    def forward(self, input):
        b, c, h, w = input.size()
        F = input.view(b, c, h*w)
        G = torch.bmm(F, F.transpose(1,2))
        return G
```

## 2단계: 이미지 로드 및 전처리

```python
def load_image(img_path):
    image = Image.open(img_path).convert('RGB')
    image = transform(image).unsqueeze(0).to(device)
    return image

img_paths = ['./네이버웹툰 사진/0000_758037.jpg','./네이버웹툰 사진/0001_648419.jpg'] # './네이버웹툰 사진/0001_648419.jpg', './네이버웹툰 사진/0002_183559.jpg'
images = [load_image(path) for path in img_paths]

# 이미지 확인
for img in images:
    plt.imshow(img.squeeze(0).permute(1, 2, 0).cpu())
    plt.show()
```

```python
def segment_image(image):
    with torch.no_grad():
        prediction = maskrcnn(image)
    masks = prediction[0]['masks']
    mask = masks[0]
    segmented_image = image * mask
    return segmented_image

segmented_images = [segment_image(img) for img in images]

# 세그멘트된 이미지 확인
for seg_img in segmented_images:
    plt.imshow(seg_img.squeeze(0).permute(1, 2, 0).cpu())
    plt.show()
```

## 여러 이미지의 bounding box 확인

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from torchvision.models.detection import maskrcnn_resnet50_fpn
import torchvision.transforms as T
import torch

# 모델 불러오기
model = maskrcnn_resnet50_fpn(pretrained=True)
model = model.eval()
```

```python
# 이미지 경로 리스트 생성
# img_paths = ['./네이버웹툰 사진/0000_758037.jpg','./네이버웹툰 사진/0001_648419.jpg','./네이버웹툰 사진/0002_183559.jpg']

import os
import random

# 이미지가 있는 폴더 경로
folder_path = './네이버웹툰 사진'

# 폴더 내의 모든 jpg 파일 리스트 가져오기
all_images = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.jpg')]

# 리스트에서 랜덤하게 20개의 이미지 선택
selected_images = random.sample(all_images, 20)  # 20은 선택하려는 이미지의 수입니다. 폴더 내의 이미지 수보다 많으면 에러가 발생합니다.
```

```python
print(os.path.basename(img for img in selected_images))
```

```python
# 이미지 경로 리스트를 반복하며 각 이미지에 대해 작업
for image_path in selected_images:
    # 이미지 불러오기
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # 이미지 전처리 및 예측
    transform = T.Compose([T.ToTensor()])
    image_tensor = transform(image).unsqueeze(0)

    with torch.no_grad():
        prediction = model(image_tensor)

    # 첫 번째 이미지에 대한 예측 가져오기
    first_image_prediction = prediction[0]

    # 결과를 그리기 위한 설정
    fig, ax = plt.subplots(1, figsize=(12, 9))
    ax.imshow(image)

    # 모든 객체에 대해 반복
    for box, score in zip(first_image_prediction['boxes'], first_image_prediction['scores']):
        if score >= 0.5:  # 신뢰도 점수가 0.5 이상인 경우만 표시
            # bounding box의 좌표를 실제 입력 이미지의 크기에 맞게 조정
            box[0] = (box[0] * original_width) / processed_size[1]  # x_min
            box[1] = (box[1] * original_height) / processed_size[0]  # y_min
            box[2] = (box[2] * original_width) / processed_size[1]  # x_max
            box[3] = (box[3] * original_height) / processed_size[0]  # y_max

            # 조정된 좌표로 bounding box 그리기
            rect = patches.Rectangle((box[0], box[1]), box[2] - box[0], box[3] - box[1], linewidth=1, edgecolor='r', facecolor='none')
            ax.add_patch(rect)

            # bounding box 위에 신뢰도 점수 표시
            ax.text(box[0], box[1], f"{score:.2f}", color='white', bbox=dict(facecolor='red', alpha=0.5))

    plt.imshow(image)
    plt.xlabel(os.path.basename(img_path))  # 이미지 아래에 파일명 표시
    plt.show()
```

### 랜덤하게 이미지 20개를 가져온 후, 1열에 신뢰도 0.5이상을 박스처리, 2열에 segment된 이미지

```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cv2
from torchvision.models.detection import maskrcnn_resnet50_fpn
import torch

# 이미지 폴더 경로
folder_path = './네이버웹툰 사진'

# 폴더 내의 모든 jpg 이미지를 가져옵니다.
all_images = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.jpg')]

# 20개의 이미지를 랜덤하게 선택합니다.
selected_images = random.sample(all_images, 20)

# Mask R-CNN 모델을 불러옵니다.
model = maskrcnn_resnet50_fpn(pretrained=True)
model.eval()

def segment_image(image, masks):
    segmented_images = []
    for mask in masks:
        mask = (mask[0] >= 0.5).astype(int)  # 마스크의 픽셀 값을 이진화
        mask = np.expand_dims(mask, axis=-1)  # 마스크의 shape을 (height, width, 1)로 변경
        segmented_image = image * mask  # 이제 shape이 맞아서 곱할 수 있습니다.
        segmented_images.append(segmented_image)
    return segmented_images

for img_path in selected_images:
    # 선택된 이미지의 파일명 출력
    print(os.path.basename(img_path))

    # 이미지 불러오기
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Mask R-CNN을 사용하여 이미지에서 객체를 검출합니다.
    input_image = torch.tensor(image.transpose((2, 0, 1)), dtype=torch.float32).unsqueeze(0) / 255.0  # Normalize
    with torch.no_grad():
        prediction = model(input_image)

    fig, axs = plt.subplots(1, 2, figsize=(6, 3))  # 1행 2열의 subplot 생성

    # 1열: 원본 이미지와 바운딩 박스
    axs[0].imshow(image)
    axs[0].set_xlabel(os.path.basename(img_path))  # 이미지 아래에 파일명 표시
    for box, score in zip(prediction[0]['boxes'].cpu().numpy(), prediction[0]['scores'].cpu().numpy()):
        if score >= 0.5:  # 신뢰도가 0.5 이상인 박스만 표시
            rect = patches.Rectangle((box[0], box[1]), box[2] - box[0], box[3] - box[1], linewidth=1, edgecolor='r', facecolor='none')
            axs[0].add_patch(rect)
            axs[0].text(box[0], box[1], f"{score:.2f}", color='white', bbox=dict(facecolor='red', alpha=0.5))  # 신뢰도 점수 표시

    # 실제 Mask R-CNN에게 선택된 바운딩 박스를 파란색으로 표시
    for mask, score in zip(prediction[0]['masks'].cpu().numpy(), prediction[0]['scores'].cpu().numpy()):
        if score >= 0.5:  # 신뢰도가 0.5 이상인 마스크만 표시
            mask = mask[0] >= 0.5  # 마스크의 픽셀 값을 이진화
            contours, _ = cv2.findContours(mask.astype('uint8'), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            for contour in contours:
                rect = cv2.boundingRect(contour)
                rect_patch = patches.Rectangle((rect[0], rect[1]), rect[2], rect[3], linewidth=1, edgecolor='b', facecolor='none')
                axs[0].add_patch(rect_patch)

    # 2열: Mask R-CNN을 통과한 세그먼트된 이미지
    segmented_images = segment_image(image, prediction[0]['masks'].cpu().numpy())
    for seg_img in segmented_images:
        axs[1].imshow(seg_img, alpha=0.5)

    plt.tight_layout()
    plt.show()
```

### 위의 방법 cosine 유사도가 너무 높게나옴
### 이 방법 => segment된 영역끼리만 비교(0이 아닌 pixel값끼리만) (20개만)

```python
import os
import torch
import torchvision.transforms as transforms
import torchvision.models as models
import torchvision.models.detection as detection
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import random

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

transform = transforms.Compose([
    transforms.ToTensor(),
])

maskrcnn = detection.maskrcnn_resnet50_fpn(pretrained=True)
maskrcnn = maskrcnn.to(device)
maskrcnn.eval()

resnet = models.resnet50(pretrained=True).to(device)
resnet = torch.nn.Sequential(*list(resnet.children())[:-1])

class GramMatrix(torch.nn.Module):
    def forward(self, input):
        b, c, h, w = input.size()
        F = input.view(b, c, h*w)
        G = torch.bmm(F, F.transpose(1,2))
        return G

def load_image(img_path):
    image = Image.open(img_path).convert('RGB')
    image = transform(image).unsqueeze(0).to(device)
    return image

folder_path = './네이버웹툰 사진'
all_img_paths = [os.path.join(folder_path, img) for img in os.listdir(folder_path) if img.endswith('.jpg')]
# all_img_paths에서 랜덤하게 20개의 이미지 경로를 선택
img_paths = random.sample(all_img_paths, 20)

images = [load_image(path) for path in tqdm(img_paths, desc='Loading Images')]

def segment_image(image):
    with torch.no_grad():
        prediction = maskrcnn(image)
    masks = prediction[0]['masks']
    if masks.shape[0] == 0:  # 세그먼트가 없을 경우
        return image, torch.zeros_like(image[0, 0]).bool()  # 원본 이미지와 빈 마스크를 반환
    mask = masks[0] >= 0.5  # 이진 마스크 생성
    segmented_image = image * mask
    return segmented_image, mask


segmented_images = []
masks = []
for img in tqdm(images, desc='Segmenting Images'):
    seg_img, mask = segment_image(img)
    segmented_images.append(seg_img)
    masks.append(mask)

def extract_features(segmented_image, mask):
    mask = mask.expand_as(segmented_image)  # 마스크의 모양을 세그먼트된 이미지의 모양과 일치시킴
    non_zero_pixels = segmented_image[mask]

    # 마스크된 영역의 픽셀들을 모아 새로운 이미지를 만듭니다.
    height, width = mask.squeeze(0).sum(0).nonzero(as_tuple=True)
    new_image = segmented_image[:, :, min(height):max(height), min(width):max(width)]

    # 마스크된 영역의 크기가 너무 작으면 원본 이미지를 사용합니다.
    if new_image.size(2) < 2 or new_image.size(3) < 2:
        feature = resnet(segmented_image)
    else:
        feature = resnet(new_image)

    gram_matrix = GramMatrix()(feature)
    return gram_matrix.cpu().data.numpy().reshape(-1)

features = [extract_features(seg_img, mask) for seg_img, mask in tqdm(zip(segmented_images, masks), desc='Extracting Features')]

def find_similar_images(query_feature, features, img_paths, top_k=10):
    similarities = cosine_similarity(query_feature.reshape(1, -1), features)
    sorted_indices = np.argsort(similarities[0])[::-1]
    similar_images = [(img_paths[i], similarities[0][i], i) for i in sorted_indices[:top_k]]  # 인덱스 i도 저장
    return similar_images

query_index = 0
query_feature = features[query_index]

similar_images = find_similar_images(query_feature, features, img_paths)

plt.imshow(images[query_index].squeeze(0).permute(1, 2, 0).cpu())
plt.title('./네이버웹툰 사진/0000_758037.jpg')
plt.show()

for img_path, similarity, idx in similar_images:  # 인덱스 idx를 가져옴
    seg_img = segmented_images[idx].squeeze(0).permute(1, 2, 0).cpu()  # 세그먼트된 이미지를 가져옴
    plt.imshow(seg_img)
    plt.title(f'Similarity: {similarity:.4f}, Filename: {os.path.basename(img_path)}')
    plt.show()
```

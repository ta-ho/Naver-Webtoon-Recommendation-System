```python
import cv2
import pytesseract

# 이미지 불러오기
image = cv2.imread('./네이버웹툰 사진/0019_805907.jpg')
```

```python
# 이미지의 높이와 너비 가져오기
h_img, w_img, _ = image.shape

# OCR 적용하여 텍스트 영역 찾기
boxes = pytesseract.image_to_boxes(image)
for b in boxes.splitlines():
    b = b.split()
    x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
    # 영역의 좌표를 정확하게 참조
    y1 = h_img - h
    y2 = h_img - y
    x1 = x
    x2 = w

    # 영역이 비어 있지 않은 경우에만 흰색으로 채우기
    if x1 < x2 and y1 < y2:
        image[y1:y2, x1:x2] = [255, 255, 255]  # 흰색으로 채우기
```

```python
import matplotlib.pyplot as plt

# 이미지 출력
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # 축 정보 없애기
plt.show()
```

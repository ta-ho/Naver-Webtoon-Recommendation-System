```python
import pandas as pd
import matplotlib as plt

df = pd.read_csv('술담화 크롤링(part1-4) 365.csv', encoding='cp949')
df 
```

### 직접 데이터를 수집하다보니? 우측 끝에 빈 열이 생김 => 제거

```python
df = df.drop(columns='Unnamed: 9')
df
```

```python
df['설명'] = df['설명'].str.replace('\n', ' ')
df['설명'] = df['설명'].str.replace('  ', '')
df['설명']
```

```python
for row in df['향미그래프']:
    print(row)
    a = row.split('|')
    print(a)
```

```python
# 모든 가능한 키들을 저장하는 집합 생성
all_keys = set()
for row in df['향미그래프']:
    if not isinstance(row, str):  # 문자열이 아닌 경우 처리
        row = '없음'
    items = row.split('|')
    for item in items:
        key = item.split(':')[0]
        all_keys.add(key)

# 각 키에 대해 데이터프레임에 새로운 열 추가
for key in all_keys:
    df[key] = '없음'

# 데이터프레임의 각 행을 순회하며 값을 할당
for idx, row in df.iterrows():
    items = row['향미그래프'].split('|')
    for item in items:
        parts = item.split(':')
        if len(parts) == 2:  # ':'로 분리했을 때 2개의 값이 반환되는 경우만 처리
            key, value = parts
            df.at[idx, key] = value

df1 = df.drop(columns=['향미그래프'])
df1
```
